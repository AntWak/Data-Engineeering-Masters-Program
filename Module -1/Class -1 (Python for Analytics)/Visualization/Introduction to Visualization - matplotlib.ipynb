# Basic of Visulization with Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib notebook 
```

```python
fig=plt.figure()
```

```python
ax1 = fig.add_subplot(2, 2, 1)
```

```python
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
```

```python
data=np.random.randn(50,).cumsum() 
```

```python
ax1.plot(data, 'k--')
```

```python
ax2.plot(data, "r-")
ax3.plot(data, "b.")
```

## Adding Legends

```python
fig=plt.figure()
```

```python
ax=fig.add_subplot(1,1,1)
```

```python
data1=np.random.randn(1000).cumsum()
data2=np.random.randn(1000).cumsum()
data3=np.random.randn(1000).cumsum()
```

```python
ax.plot(data1, 'k', label='one')
ax.plot(data2, 'b--', label='two')
ax.plot(data3, 'r.', label='three')
```

```python
ax.legend()
```
