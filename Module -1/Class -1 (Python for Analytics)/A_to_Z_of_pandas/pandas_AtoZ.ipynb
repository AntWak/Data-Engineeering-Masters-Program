### Imputation Method 

There are many options we could consider when replacing a missing value, for example:

* constant value that has meaning within the domain, such as 0, distinct from all other values.
* value from another randomly selected record.
* mean, median or mode value for the column.
* value estimated by another predictive model.
* Imterpolation methods like
    * linear
    * quadratic
    * polynomial
    * time series etc

   <ul>

    <li><b>Applying</b>
        <ul>
            <li>Aggregation</li>
            <li>Transformation</li>
            <li>Filterting</li>
        </ul>
    </li>
    <li><b>Combine</b></li>
   </ul>
        
   Groupby property is grouped the data according the column supplied to the function.In the following example

### Concatenation of DataFrame 
------------
    concatanation of dataframes in -
          1)Append by row
          2)Append by col
  

```python
temp1={
      "date":['01-02-12','03-02-12','04-02-12','05-02-12'],
      "event":['sunny','cold','cold','rainy'],
      "temp":[14,16,15,10]
}
temp=pd.DataFrame(temp1)
temp
```

```python
ws={
      "date":['01-02-12','03-02-12','04-02-12','05-02-12'],
      "event":['sunny','cold','cold','rainy'],
      "wind-speed":[12,10,9,14],
}
wind_speed=pd.DataFrame(ws)
wind_speed
```

```python
df=pd.concat([temp,wind_speed],axis=1)
df.head()
```

```python
#Incorrect as it has been concatenated by rows
df=pd.concat([temp,wind_speed],axis=0)
df.head()
```

#### Merging dataframes 
    merge() is a method in pandas in which you can merge two dataframes 
#### Joins
    * inner
    * outer
    * right
    * left

```python
product=pd.DataFrame({
    'P_ID':[1001,1002,1003,1004,1005,1006,1007],
    'P_name':['W','B','Sh','Sm','Bo','O','L'],
    'Cat':['F','F','Fashion','E','S','G','E'],
    'Price':[199.0,1230.50,2789.0,15699.0,155.0,140.0,78899.0],
    'Sell_City':['Delhi','Mumbai','Chennai','Kolkata','Delhi','Chennai','Bengalore']
})
```

```python
customer=pd.DataFrame({
    'id':[1,2,3,4,5,6,7,8,9],
    'Name':['Olivia','Aditya','Cory','Isabell','Dominic','Tyler','Samuel','Daniel','Jeremy'],
    'Age':[18,27,25,40,20,55,34,17,33],
    'P_ID':[1001,0,1006,0,1003,1004,0,0,1007],
    'PurchasedProduct':['W','NA','O','NA','Sh','Sm','NA','NA','L'],
    'City':['Mumbai','Delhi','Bangalore','Chennai','Chennai','Delhi','Kolkata','Delhi','Mumbai']
})
```

```python
product.head()
```

```python
customer.head()
```

```python
#Merging dataframes with the concept of inner join
pd.merge(product,customer,on='P_ID')
```

```python
# Inner join on Product_name and Purchased_Product
pd.merge(product,customer,left_on='P_name',right_on='PurchasedProduct')
```

```python
pd.merge(product,customer,on='P_ID',how='outer')
```

```python
pd.merge(product,customer,on='P_ID',how='left')
```

```python
pd.merge(product,customer,on='P_ID',how='right')
```

### Pivot and Pivot table

* Pivot allows you reshape or trandform your dataset.

### CrossTab

```python
# Load necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
pd.crosstab(index=tips['sex'], columns=tips['smoker'])
```

```python
df=pd.crosstab(index=tips['day'],
            columns=tips['time'],
            values=tips['total_bill'],
            aggfunc=np.mean).round(0)
df
```

```python
sns.heatmap(df, cmap='rocket_r', annot=True, fmt='g');
```

### Reshaping Dataframe 
    Reshaping the dataframe with melt() method.

```python
mydis={
    'Day':['Mon','Tue','Wed','Thu','Fri','Sat','Sun'],
    'Kolkata':[10,20,30,40,32,34,23],
    'Kerala':[23,32,13,42,13,43,23],
    'Mumbai':[10,20,30,40,32,34,23]
}
df=pd.DataFrame(mydis)
df
```

```python
df2=pd.melt(df,id_vars=['Day'],var_name='City',value_name='Temp')
df2
```
