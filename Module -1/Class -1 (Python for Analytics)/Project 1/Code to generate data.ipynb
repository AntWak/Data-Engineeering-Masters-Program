# Importing source file and removing dirty records as an initial cleaning

```python
#Importing source file without header
df1 = pd.read_csv("billings_europe.csv", header=None)
df1.head()
```

```python
# Removing first 4 rows as a first step towards removing irrelavant data (Dirty Data)

df2 = df1.drop([df1.index[0] , df1.index[1], df1.index[2],df1.index[3]])
df2.head()
```

```python
# Analysing unique records in 2nd column (Column =1 )
df2[1].unique()
```

```python
# Removing column 1 as it contains only null values

df3=df2.drop([1], axis=1)
df3.head()
```

# Transposing the dataframe and starting with data wrangling

```python
#Transposing data for a better view

df4 = df3.transpose()
df4.head()
```

**Data Cleansing**

```python
# Checking the total number of null values per column
df4.isnull().sum()
```

```python
# Analysing the values of a records at particular position (an example where all the column values are NULL)

df4.iloc[52]
```

```python
#Dropping the records which have all the column values as NULL

df5=df4.dropna(how='all')
df5.head()
```

### If the Type is missing which is Column 5 in dataframe df5, taking the last non-missing value to the top (As per suggestions in problem statement)
### Considering non-missing value from the top as we have taken transpose of original dataframe.

```python
# Analysing 1st record with missing value of Type (Column 5)

df5.iloc[4]
```

```python
# Using the concept of forward fill method from Pandas to fill missing values in Type

df6 = df5
df6[5] = df6[5].fillna(method="ffill")
df6.head()
```

```python
#Checking if the value populated correctly for the  1st record with missing value of Type (Column 5) [Value populated correctly]

df6.iloc[4]
```

```python
df6.head()
```

```python
#Relacing the missing value with "" in order to perform few operations on Column 4 like spliting the value and applying lambda function to segrregate Segment and Period.
df6[4]=df6[4].fillna("")
df6.head()
```

#### Creating new column 3 for Segment and applying split function to segrregate value on string " - " and applying lambda function to populate the value throughout the column.


```python
df6[3]=df6[4].apply(lambda x: (x.split(' - ',2)[0]))
df6.head()
```

```python
#Sorting the dataframe on the basis of Column

df6.reset_index(drop=True)
df6.sort_index(axis=1, inplace = True)
```

```python
df6.head()
```

```python
# Resetting the index value

df6.reset_index(drop=True)
```

```python
# Adding the future column names of Data Frame at (0,0),(0,1),(0,2) and (0,3)

df6.iat[0,0]='Segment'
df6.iat[0,1]='Period'
df6.iat[0,2]='Type'
df6.iat[0,3]='Subtype'
```

```python
df6.head()
```

```python
#Renaming Column names to 1st row data and dropping row at index = 0

df7=df6.rename(columns=df6.iloc[0]).drop(df6.index[0])
df7.head()
```

```python
# Applying reverse pivot_table concept - Reshaping the data

df8=pd.melt(df7,id_vars=["Segment","Period","Type","Subtype"])
df8.head()
```

```python
# Renaming the columns as per Problem statement

df8.columns = ['Segment', 'Period', 'Type', 'Subtype', 'Date', 'Value']
```

```python
#Verifying column names

list(df8.columns)
```

```python
# Sorting the data frame on column index

df8.sort_index(axis=1, inplace = True)
df8
```

```python
df8.reset_index(drop=True)
```

```python
#Creating new dataframe with customised position of columns

df9=df8[['Date','Segment','Period','Type','Subtype','Value']]
df9.head()
```

```python
dfx=df8[['Date','Segment','Period','Type','Subtype','Value']]
```

```python
df9.head()
```

```python
# Sorting the Data by Segment and Date 

df9.sort_values(by=['Segment','Date'], inplace = True)
df9.head()
```

```python
df10=df9
df10
```

```python
# Updating the values of Period as per problem statement. Applied split function to segrregate value on string " - " 
# and applied lambda function to populate the value at postion 1 throughout the column.

df10['Period']=df10['Period'].apply(lambda x: (x.split(' - ',2)[1]))
```

```python
df10
```

```python
# Removing all trailing/leading spaces if any from all the columns

df10 = df10.applymap(lambda x: x.strip() if isinstance(x, str) else x)
df10.head()
```

```python

df10.isnull().sum()
```
